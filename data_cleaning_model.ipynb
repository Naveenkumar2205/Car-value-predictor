```python
#importing necessary libraries
import pandas as pd
import numpy as np
```

```python
#reading data
data = pd.read_excel('C:/Users/navee/OneDrive/Desktop/Book1.xlsx')
data.head()
```

```python
#keeping only required columns
data_keep = data[['year','make','model','listing_mileage','dealer_state','dealer_city','listing_price']]
data_keep.head()
```

```python
#removing null values
data_keep.dropna(axis=0,inplace=True)
```

```python
#checking if there are any null values
data_keep.isnull().sum()
```

```python
data_keep.head()
```

```python
print(data_keep['dealer_city'].value_counts())
print(data_keep['dealer_state'].value_counts())
#since dealer city is having more number of categorical variables, it is removed
```

```python
#fidning out the count value of car names
counts=data_keep['make'].value_counts()
print(counts)

#finding all the car_names that are having count above 5000
data_keep1 = data_keep[data_keep['make'].map(counts > 5000)]
data_keep1
```

```python
#fidning out the count value of car names
counts1=data_keep1['model'].value_counts()
print(counts1)

#finding all the car_names that are having count above 1000
data_keep2 = data_keep1[data_keep1['model'].map(counts1 > 1000)]
data_keep2
```

```python
#now removing dealer_city
data_keep2.drop('dealer_city',axis=1,inplace=True)
```

```python
data_keep2
```

```python
data_keep2['make']=data_keep2['make'].astype('object')
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# Convert the 'make' , 'model' and 'dealer_state' columns to string data type
data_keep2['make']=data_keep2['make'].astype(str)
data_keep2['model'] = data_keep2['model'].astype(str)
data_keep2['dealer_state'] = data_keep2['dealer_state'].astype(str)

# Fit and transform the 'make' column
data_keep2['make'] = data_keep2['make'].str.upper()
make_ini = data_keep2['make']

data_keep2['make'] = le.fit_transform(data_keep2['make'])

make_fi = data_keep2['make']
mapping = dict(zip(make_ini, make_fi))
print(mapping)

 #Fit and transform the 'model' column
model_ini = data_keep2['model']

data_keep2['model'] = le.fit_transform(data_keep2['model'])

model_fi = data_keep2['model']
mapping1 = dict(zip(model_ini, model_fi))
print(mapping1)

# Fit and transform the 'dealer_state' column
dealer_state_ini= data_keep2['dealer_state']

data_keep2['dealer_state'] = le.fit_transform(data_keep2['dealer_state'])

dealer_state_fi= data_keep2['dealer_state']
mapping2= dict(zip(dealer_state_ini, dealer_state_fi))
print(mapping2)

# Print the updated DataFrame
data_keep2
```

```python
data_keep2.to_csv('model_prediction1.csv')
```
